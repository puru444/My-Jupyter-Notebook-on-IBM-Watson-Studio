<b>2. How to make Bulleted List in HTML?</b>

HTML lists allow web developers to group a set of related items in lists. There are various types of lists have been used in HTML however there're mainly 2 types of lists have been used frequently:

<ul>
    <li>Ordered Lists</li>
    <li>Unordered Lists</li>
</ul>

<b>Ordered List:</b> An ordered list also known as "Numbered List" starts with the < ol > tag. Each list item starts with the < li > tag. By putting < ol >, the list items will be marked with numbers by default.

Example:
The first 4 months of any year comes in this order:
 <ol>
  <li>January</li>
  <li>February</li>
  <li>March</li>
  <li>April</li>
</ol> 

<b>Unordered List:</b> An unordered list also known as "Bulleted List" starts with the < ul > tag. Each list item starts with the < li > tag. By putting < ul >, the list items will be marked with bullets (small black circles) by default.

Example:
The 3 main products of Apple are:
 <ul>
  <li>Macbook</li>
  <li>iPad</li>
  <li>iPhone</li>
</ul>

<b>3. Tables in HTML.</b>

A table in HTML consists of table cells inside rows and columns. Where, following are the main attributes of table:

<dl>
  <dt>td tag</dt>
  <dd>- Each table cell is defined by a < td > and a < /td > tag</dd>
  <dd>- And, td stands for table data</dd>
</dl>
    
<dl>  
  <dt>tr tag</dt>
  <dd>- Each table row starts with a < tr > and ends with a < /tr > tag.</dd>
  <dd>- And, tr stands for table row</dd>
</dl>   

<dl>
  <dt>th tags</dt>
  <dd>- Sometimes we want our cells to be table header cells. In those cases we use the < th > tag instead of the < td > tag.</dd>
  <dd>- And, th stands for table header. By default, the text in < th > elements are bold and centered, but yowe can change that with CSS. </dd>
</dl>
